```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
expo_data = linear_data ** 2

plt.figure() # it is called by default
plt.plot(linear_data, '-o', expo_data, '-o')
```

```python
plt.plot(linear_data, '-o', expo_data, '-o')
plt.plot([22,44,55], '--r')
```

```python
plt.plot(linear_data, '-o', expo_data, '-o')
plt.gca().fill_between(range(len(linear_data)), linear_data, expo_data, facecolor='blue', alpha=0.25)

plt.title("Expo ($x^2$) vs. Linear ($x$) Performance")
```

### work with dates

```python
dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(dates, linear_data, '-o', dates, expo_data, '-o')
```
