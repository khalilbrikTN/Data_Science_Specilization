```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np

linearData = np.array([1,2,3,4,5,6,7,8])

xvals = range(len(linearData))

plt.bar(xvals, linearData, width=0.3)
```

```python
newXvals = []
expoData = linearData ** 2
for item in xvals:
    newXvals.append(item+0.3)

plt.bar(xvals, linearData, width=0.3)
plt.bar(newXvals, expoData, width=0.3, color = 'r')
```

```python
from random import randint
linear_err = [randint(1,4) for x in range(len(linearData))]

plt.bar(xvals, linearData, width=0.3, yerr = linear_err)
```

```python
plt.bar(xvals, linearData, width=0.3, color = 'b')
plt.bar(xvals, expoData, width=0.3, bottom=linearData, color='r')
```

```python
plt.barh(xvals, linearData, height=0.3, color = 'b')
plt.barh(xvals, expoData, height=0.3, left=linearData, color='r')
```
