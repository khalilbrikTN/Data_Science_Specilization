# Without using the scripting layer

```python
#we will be now using the FigureCanvasAgg
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure

fig = Figure()

canvas = FigureCanvasAgg(fig) #add backend to fig

ax = fig.add_subplot(111) #the axis object

ax.plot(3,2,".")

canvas.print_png('test.png')
```
